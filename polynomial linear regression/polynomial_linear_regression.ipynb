```python
# importing libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# importing dataset
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(x)
```

```python
print(y)
```

```python
# linear regression model
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x, y)
```

```python
# polynomial regression model
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
x_poly=poly_reg.fit_transform(x)
lin_reg_2=LinearRegression()
lin_reg_2.fit(x_poly,y)
```

```python
print(x_poly)
```

```python
# visualizing linear regression result
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x))
plt.title('Truth or bluff (linear regression)')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
# visualizing polynomial regression result
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg_2.predict(poly_reg.fit_transform(x)))
plt.title('Truth or bluff (polynomial regression')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
# visualising polynomial regression result (higer resolution curve)
x_grid=np.arange(min(x),max(x),0.1)
x_grid=x_grid.reshape((len(x_grid),1))
plt.scatter(x,y,color='red')
plt.plot(x_grid,lin_reg_2.predict(poly_reg.fit_transform(x_grid)))
plt.title('Truth or bluff (polynomial regression')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
# pridicting  a new result with linear regression
lin_reg.predict([[6.5]])
```

```python
# pridicting  a new result with polynomial regression
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
